```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('campaign_metrics.csv')

# 1. Cost Per Click (CPC)
df['CPC'] = df['mark_spent'] / df['clicks']

# 2. Cost Per Thousand Impressions (CPM)
df['CPM'] = (df['mark_spent'] / df['impressions']) * 1000

# 3. Average Order Value (AOV)
df['AOV'] = df['revenue'] / df['orders']

# 4. Revenue Per Click (RPC)
df['RPC'] = df['revenue'] / df['clicks']

# Save the updated dataset
df.to_csv('Cost_Analysis.csv', index=False)

# Preview the results
print(df[['campaign_name', 'CPC', 'CPM', 'AOV', 'RPC']])
```

```python
# Set up the plot size
plt.figure(figsize=(14, 10))

df=pd.read_csv('project digital advertising/Cost_Analysis.csv')

# 1. Visualising CPC by Campaign
plt.subplot(2, 2, 1)
sns.barplot(x='CPC', y='campaign_name',hue='campaign_name', data=df, palette='Blues_d') 
plt.title('Cost Per Click (CPC) by Campaign')

# 2. Visualising CPM by Campaign
plt.subplot(2, 2, 2)
sns.barplot(x='CPM', y='campaign_name', hue='campaign_name',data=df, palette='Blues_d')
plt.title('Cost Per Thousand Impressions (CPM) by Campaign')

# 3. Visualising AOV by Campaign
plt.subplot(2, 2, 3)
sns.barplot(x='AOV', y='campaign_name',hue='campaign_name', data=df, palette='Blues_d')
plt.title('Average Order Value (AOV) by Campaign')

# 4. Visualising RPC by Campaign
plt.subplot(2, 2, 4)
sns.barplot(x='RPC', y='campaign_name',hue='campaign_name', data=df, palette='Blues_d')
plt.title('Revenue Per Click (RPC) by Campaign') 
```

Campaign Performance Analysis (CPC, CPM, AOV, RPC)
This visualisation presents four key metrics across different advertising campaigns:

Cost Per Click (CPC)
Cost Per Thousand Impressions (CPM)
Average Order Value (AOV)
Revenue Per Click (RPC)

Key Takeaways:
1. Cost Per Click (CPC)
Highest CPC: Facebook Lookalike (LAL) campaign has the highest CPC (~20+), making it one of the most expensive campaigns in terms of direct ad spend.
Lowest CPC: Instagram Tier 2 has the lowest CPC, indicating a cost-effective way to drive clicks, but conversion quality needs further examination.

2. Cost Per Thousand Impressions (CPM)
High CPM campaigns:
Facebook Retargeting and Google Wide have significantly higher CPM, meaning they are costly in terms of ad exposure.
Low CPM campaigns:
Instagram Tier 2 and Banner Partner have the lowest CPM, suggesting lower competition or broader audience reach.

3. Average Order Value (AOV)
Best-Performing Campaigns (High AOV):
YouTube Blogger and Google Hot generate the highest AOV (~7000-8000), meaning customers from these sources tend to spend more per order.
Lower AOV:
Facebook Tier 1 and Tier 2 campaigns have significantly lower AOV (~1000-2000).

4. Revenue Per Click (RPC)
High RPC (High Revenue Generation Per Click):
YouTube Blogger and Google Hot stand out with RPC values exceeding 30-40, making them the most profitable in terms of revenue per user interaction.
Lower RPC:
Instagram Tier 2 and Facebook Tier 1 campaigns generate relatively lower RPC, indicating lower immediate returns per user click.

Recommendations
Invest More in High-Performing Campaigns:
Campaigns with high AOV & RPC (YouTube Blogger, Google Hot, and Facebook Retargeting) should receive higher budgets and expansion efforts.

Reevaluate Costly Yet Inefficient Campaigns:
Facebook Lookalike (LAL) has high CPC and CPM, but unless its conversion rates justify the expense, it may need a budget adjustment or audience retargeting.

Monitor & Optimise Low-Cost Campaigns:
Instagram Tier 2 and Banner Partner have low CPC and CPM, meaning they are cheap in terms of traffic acquisition. However, their conversion rates and revenue generation should be closely monitored to avoid wasted spend.

Final Thoughts
Campaigns like YouTube Blogger and Google Hot seem to deliver the best return on investment.
Facebook Retargeting is another strong performer in AOV and conversion rates, making it an essential part of the funnel.
Budget allocations should be optimised towards campaigns that balance cost-effectiveness with high revenue generation.
